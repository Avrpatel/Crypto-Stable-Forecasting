```python
# Combine dataframes

busd_all_years = [busd_df_1, busd_df_2, busd_df_3]
dai_all_years = [dai_df_1, dai_df_2, dai_df_3]
usdc_all_years = [usdc_df_1, usdc_df_2, usdc_df_3]
usdt_all_years = [usdt_df_1, usdt_df_2, usdt_df_3]

busd_df_all= pd.concat(busd_all_years)
dai_df_all= pd.concat(dai_all_years)
usdc_df_all= pd.concat(usdc_all_years)
usdt_df_all= pd.concat(usdt_all_years)
```

```python
def clean_format(df):
  df = df.drop(columns=['perc_chg','color'])
  df['date'] = pd.to_datetime(df['date'], unit='s')
  df = df.sort_values(by='date', ascending=True)
  df = df.dropna()
  df["vol"] = df["vol"].replace({"K":"*1e3", "M":"*1e6", "B":"*1e9"}, regex=True).map(pd.eval).astype(float)
  df = df.astype({'price':'float64','open':'float64','high':'float64','low':'float64'})
  
  return df

```

```python
# Drop the unnecessary columns 

busd_df_all = busd_df_all.drop(columns=['perc_chg','color'])
dai_df_all = dai_df_all.drop(columns=['perc_chg','color'])
usdc_df_all = usdc_df_all.drop(columns=['perc_chg','color'])
usdt_df_all = usdt_df_all.drop(columns=['perc_chg','color'])
```

```python
# Drop duplicate dates

busd_df_clean = busd_df_all.drop_duplicates(subset=['date'])
dai_df_clean = dai_df_all.drop_duplicates(subset=['date'])
usdc_df_clean = usdc_df_all.drop_duplicates(subset=['date'])
usdt_df_clean = usdt_df_all.drop_duplicates(subset=['date'])
```

```python
# Drop nulls

busd_df_clean = busd_df_clean.dropna()
dai_df_clean = dai_df_clean.dropna()
usdc_df_clean = usdc_df_clean.dropna()
usdt_df_clean = usdt_df_clean.dropna()
```

```python
# Set index

busd_df_clean = busd_df_clean.set_index('date')
dai_df_clean = dai_df_clean.set_index('date')
usdc_df_clean = usdc_df_clean.set_index('date')
usdt_df_clean = usdt_df_clean.set_index('date')
```

```python
# Cleaning/Changing Data Types

busd_clean = busd_df_clean
dai_clean = dai_df_clean
usdc_clean = usdc_df_clean
usdt_clean = usdt_df_clean

busd_clean["vol"] = busd_clean["vol"].replace({"K":"*1e3", "M":"*1e6", "B":"*1e9"}, regex=True).map(pd.eval).astype(float)
dai_clean["vol"] = dai_clean["vol"].replace({"K":"*1e3", "M":"*1e6", "B":"*1e9"}, regex=True).map(pd.eval).astype(float)
usdc_clean["vol"] = usdc_clean["vol"].replace({"K":"*1e3", "M":"*1e6", "B":"*1e9"}, regex=True).map(pd.eval).astype(float)
usdt_clean["vol"] = usdt_clean["vol"].replace({"K":"*1e3", "M":"*1e6", "B":"*1e9"}, regex=True).map(pd.eval).astype(float)
```

```python
# Changing other data types 

busd_clean = busd_clean.astype({'price':'float64','open':'float64','high':'float64','low':'float64'})
dai_clean = dai_clean.astype({'price':'float64','open':'float64','high':'float64','low':'float64'})
usdc_clean = usdc_clean.astype({'price':'float64','open':'float64','high':'float64','low':'float64'})
usdt_clean = usdt_clean.astype({'price':'float64','open':'float64','high':'float64','low':'float64'})
```

```python
def clean_split_train(df):
    df = df.sort_index(ascending=True)
    ml_df = df.reset_index()

    length = ml_df.shape[0]
    split_point = round(length*.70)

    train_df = ml_df.loc[:split_point]

    return train_df
```

```python
def clean_split_test(df):
    df = df.sort_index(ascending=True)
    ml_df = df.reset_index()

    length = ml_df.shape[0]
    split_point = round(length*.70)

    test_df = ml_df.loc[split_point+1:]

    return test_df
```
